# SandBox CLS-FR



```python
from SCBert.load_data import DataLoader

cls = DataLoader().load_cls_fr().sample(frac=1)
```

```python
from SCBert.SCBert import Vectorizer

vectorizer = Vectorizer("flaubert")
```

```python
text_vectors = vectorizer.vectorize(cls.review.iloc[:500], layers=[10,11,12], word_pooling_method="average", sentence_pooling_method="average")
```

```python
from SCBert.SCBert import EmbeddingExplorer

ee = EmbeddingExplorer(cls.review[:500],text_vectors)
```

```python
labels = ee.cluster(3)
```

```python
ee.extract_keywords()
```

```python
ee.explore(cls.code.iloc[:500])
```
