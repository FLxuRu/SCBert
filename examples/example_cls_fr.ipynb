# SandBox CLS-FR

Just a playground to test all the functions available now (version 0.0.3) 

```python
from SCBert.load_data import DataLoader

cls = DataLoader().load_cls_fr().sample(frac=1)
```

```python
from SCBert.SCBert import Vectorizer

vectorizer = Vectorizer("flaubert")
```

```python
data = cls.review.iloc[:500]
```

```python
text_vectors = vectorizer.vectorize(data, layers=[10,11,12], word_pooling_method="average", sentence_pooling_method="average")
```

Explore the data to find groups that make sense with our vector computer in the last step. 

```python
from SCBert.SCBert import EmbeddingExplorer

ee = EmbeddingExplorer(data,text_vectors)
```

```python
labels = ee.cluster(3)
```

```python
ee.extract_keywords()
```
